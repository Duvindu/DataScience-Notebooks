```python
from sklearn.manifold import TSNE
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import torchvision
import torchvision.datasets as datasets
```

# Data Preparation

```python
mnist_trainset = datasets.MNIST(root='./data', train=True, download=True, transform=None)
# mnist_testset  = datasets.MNIST(root='./data', train=False, download=True, transform=None)

print(len(mnist_trainset))
print(len(mnist_testset))

train_im_list    = np.zeros((len(mnist_trainset), (28* 28)))
train_label_list = np.zeros(len(mnist_trainset), dtype=int)
                            
for idx, data in enumerate(mnist_trainset):
    train_im, train_label = data
    train_im_list[idx]    = np.reshape(train_im, (28*28))
    train_label_list[idx] = int(train_label)

# print(train_im_list[0])
```

# t-SNE visualization

```python
def plot_scatter(values, cls):
    num_classes = 10
    # Create a color-map with a different color for each class.
    cmap = cm.rainbow(np.linspace(0.0, 1.0, num_classes))
    # Get the color for each sample.
    print(cls[0:10])
    colors = cmap[cls]
    # Extract the x- and y-values.
    x = values[:, 0]
    y = values[:, 1]
    # Plot it.
    plt.scatter(x, y, color=colors)
    plt.show()

tsne = TSNE(n_components=2)

tsne_fitted_data = tsne.fit_transform(train_im_list[1:-1:50])
print(tsne_fitted_data.shape)
plot_scatter(tsne_fitted_data, train_label_list[1:-1:50])
```
